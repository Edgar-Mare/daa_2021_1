```
import numpy as np

td=[]
for i in range(1001):
  fila=[]
  fila.append(i)
  fila.append(np.random.normal(15,4))
  fila.append(np.random.randint(1,6))
  td.append(fila)

pila=[]
tamanoPila=12000
sumaTamanos=0
prioridad=5
i=0

while sumaTamanos<=tamanoPila and prioridad>0:
  if td[i][2]==prioridad:
    if (sumaTamanos+td[i][1])<=tamanoPila:
      pila.append(td[i])
      sumaTamanos+=td[i][1]
  if i>=len(td)-1:
    prioridad-=1
    i=0
  i+=1

print(pila)
print(len(pila))
suma=0
for j in range(len(pila)):
  suma+=pila[j][1]
print(suma)
```
